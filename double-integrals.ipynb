## Integral Bounded by Numbers ##

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# Create a 2D meshgrid for x and y values
x = np.linspace(0, 2, 100)
y = np.linspace(0, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the function f(x, y) in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Create a contour plot to highlight the area for the integral
contour_levels = [f(0, 0), f(2, 0), f(2, 3)]
CS = ax.contourf(X, Y, Z, contour_levels, zdir='z', offset=f(0, 0), cmap='Blues', alpha=0.5)

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('f(x, y) = x^2 + y^2')

# Show the 3D plot
plt.show()
```

## Integral Bounded by Simple Curves ##

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function g(x, y) = x + y
def g(x, y):
    return x + y

# Define the vertices of the triangular region T
vertices = np.array([[0, 0], [1, 0], [1, 1]])

# Create a 2D meshgrid for x and y values in the triangular region T
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Mask the points outside the triangular region T
mask = (Y <= X) & (Y >= 0) & (X <= 1)
Z = np.where(mask, g(X, Y), np.nan)

# Plot the function g(x, y) in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Color the triangular region T in light blue to represent the integral
ax.plot_trisurf(vertices[:, 0], vertices[:, 1], [g(*vertex) for vertex in vertices], color='lightblue', alpha=0.5)

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('g(x, y)')
ax.set_title('g(x, y) = x + y')

# Show the 3D plot
plt.show()
```
